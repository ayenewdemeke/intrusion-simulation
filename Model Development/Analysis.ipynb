**Import required libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
import numpy as np
```

**Load the data from the CSV file**

```python
# Load the data from the CSV file
file_path = 'small_car.csv'
data = pd.read_csv(file_path)
```

**Extract the columns for analysis**

```python
speed_data = data['Speed (km/hr)']
acceleration_data = data['Max Acceleration (m/s^2)']
```

**Plotting the distribution of speed and acceleration**

```python
plt.figure(figsize=(12, 6))

# Speed
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
sns.histplot(speed_data, stat="density", linewidth=0, color="blue")
mean_speed, std_speed = norm.fit(speed_data)
xmin_speed, xmax_speed = plt.xlim()
x_speed = np.linspace(xmin_speed, xmax_speed, 100)
p_speed = norm.pdf(x_speed, mean_speed, std_speed)
plt.plot(x_speed, p_speed, 'k', linewidth=2)
plt.title("Speed: mean = {:.2f}, std = {:.2f}".format(mean_speed, std_speed))
plt.xlabel('Speed (km/hr)')
plt.ylabel('Density')
plt.grid(True)

# Acceleration
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
sns.histplot(acceleration_data, stat="density", linewidth=0, color="red")
mean_accel, std_accel = norm.fit(acceleration_data)
xmin_accel, xmax_accel = plt.xlim()
x_accel = np.linspace(xmin_accel, xmax_accel, 100)
p_accel = norm.pdf(x_accel, mean_accel, std_accel)
plt.plot(x_accel, p_accel, 'k', linewidth=2)
plt.title("Acceleration: mean = {:.2f}, std = {:.2f}".format(mean_accel, std_accel))
plt.xlabel('Max Acceleration (m/s^2)')
plt.ylabel('Density')
plt.grid(True)

plt.tight_layout()
plt.show()
```
